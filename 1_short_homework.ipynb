```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wavfile
import statistics

Fs, aud = wavfile.read('/content/1kis-hf/PinkPanther30.wav')
def show_info(aname, a):
    print ("Array:", aname)
    print ("min, max:", a.min(), a.max())
    print

show_info("Original", aud)
print ("expectedValueBefore:",np.mean(aud))
print ("stDevBefore:",np.std(aud))
print('\n')
zeroExpectedValue = aud - np.mean(aud)
show_info("zeroExpectedValue",zeroExpectedValue)
stDeviation = zeroExpectedValue/(statistics.stdev(zeroExpectedValue))
print("zeroExpectedValue:",np.mean(zeroExpectedValue))
print('\n')
show_info("stDeviation",stDeviation)
print("stDev:",statistics.stdev(stDeviation))
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(aud, Fs=Fs)
plt.show()


print('\n')
Fs, aud = wavfile.read('/content/1kis-hf/StarWars60.wav')
first = aud[:int(Fs*30)]
show_info("Original", aud)
print ("expectedValueBefore:",np.mean(aud))
print ("stDevBefore:",np.std(aud))
print('\n')
zeroExpectedValue = aud - np.mean(aud)
show_info("zeroExpectedValue",zeroExpectedValue)
stDeviation = zeroExpectedValue/(statistics.stdev(zeroExpectedValue))
print("zeroExpectedValue:",np.mean(zeroExpectedValue))
print('\n')
show_info("stDeviation",stDeviation)
print("stDev:",statistics.stdev(stDeviation))
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(first, Fs=Fs)#(first, Fs=Fs)
plt.show()
```

```python
import matplotlib.pyplot as plt
from urllib.request import urlopen
from bs4 import BeautifulSoup
from collections import Counter

url = "http://smartlab.tmit.bme.hu/oktatas-deep-learning"
html = urlopen(url).read()  
soup = BeautifulSoup(html, 'lxml') 
for tag in soup.find_all(['script', 'style']):
    tag.decompose()   
text = soup.get_text(strip=True)
print(text)


the_input = text
letters =Counter(char for char in the_input.lower() if char.isalpha())
print (letters)
plt.figure(figsize=(15, 6))
plt.bar(letters.keys(), letters.values(), width=0.5, color='r')
plt.show()
```
